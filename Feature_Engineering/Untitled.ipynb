```python
import seaborn as sns 
```

```python
df=sns.load_dataset("tips")
```

```python
df.head ()
```

```python
bill_mean=df['total_bill'].mean()
bill_std=df['total_bill'].std()
```

```python
bill_mean,bill_std
```

```python
bill=df["total_bill"]
```

```python
std_bill=[]
z_score=0
for i in bill:
    z_score=(i-bill_mean)/bill_std
    std_bill.append(z_score)
    
```
